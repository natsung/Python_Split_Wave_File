```python
# This example is used for split wave file data into chunks, not specifix chunks length
# Load the required libraries:
#   * scipy
#   * numpy
#   * matplotlib
#   * glob & os

from scipy.io import wavfile
from matplotlib import pyplot as plt
import numpy as np
import glob, os

wave_file_input_dir = "wave_dir/"
wave_file_output_dir = "wave_split_dir/"
split_chunks = 3

for file in os.listdir(wave_file_input_dir):
    if file.endswith(".wav"):
        wave_in = os.path.join(wave_file_input_dir, file)
        wave_out = os.path.join(wave_file_output_dir, file.replace(".wav",""),)
    
        # Load the data and calculate the time of each sample
        samplerate, data = wavfile.read(wave_in)
        #print("samplerate:",samplerate,len(data),data)

        plt.plot(data)
        plt.show()

        wavelen = len(data)
        
        split_len = wavelen / split_chunks
        #print(split_len)

        for i in xrange(1,split_chunks+1):
            chunks_data = data[split_len*(i-1):split_len*i]
            #plt.plot(chunks_data)
            #plt.show()
            wavfile.write("%s_%d.wav"%(wave_out,i), samplerate, chunks_data)
```
